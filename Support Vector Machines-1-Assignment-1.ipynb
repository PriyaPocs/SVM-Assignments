Q1. What is the mathematical formula for a linear SVM?


Answer(Q1):

The mathematical formulation for a linear Support Vector Machine (SVM) involves finding the hyperplane that best separates two classes of data points while maximizing the margin between the two classes. The basic idea is to find a hyperplane that can separate the data with the widest possible margin.

Given a set of data points (x1, y1), (x2, y2),..., (xn, yn), where xi represents the feature vectors and yi represents the corresponding class labels (yi in {-1, 1\}), the linear SVM aims to find a hyperplane described by the equation:

[ w.x + b = 0 \]

where:
- w is the weight vector perpendicular to the hyperplane.
- b is the bias term that shifts the hyperplane.

The goal is to find \(w\) and \(b\) that satisfy the following conditions:
1. For positive-class data points: w.xi + b >= 1 if yi = 1.
2. For negative-class data points: w.xi + b <= -1 if yi = -1.

These conditions ensure that data points are correctly classified and that there is a margin of at least 2/||w|| between the two classes.



To maximize the margin, the SVM aims to minimize 2/||w||^2 while satisfying the classification conditions. This leads to the following optimization problem:

Minimize: 2/||w||^2
Subject to: yi*(w.xi + b) >= 1 for all i = 1, 2, ..., n.




This is a quadratic programming problem that can be solved using various optimization algorithms. Once the optimal \(w\) and \(b\) are found, the hyperplane \(w \cdot x + b = 0\) can be used for classification. New data points are classified based on the sign of \(w \cdot x + b\).

It's important to note that in practice, most SVM implementations also introduce the notion of soft margins to handle cases where the data might not be perfectly separable. This involves introducing slack variables and modifying the optimization problem slightly to allow for some misclassification.




Q2. What is the objective function of a linear SVM?


Answer(Q2):

The objective function of a linear Support Vector Machine (SVM) is to find the parameters (weight vector \(w\) and bias \(b\)) that define a hyperplane with the maximum margin while minimizing classification errors. Mathematically, this objective can be formulated as an optimization problem.


The optimization problem seeks to minimize the square of the L2 norm |w|^2 of the weight vector while satisfying the classification constraints. This leads to finding the hyperplane that maximizes the margin between the classes, as the margin is inversely proportional to the norm of \(w\). Smaller values of ||w|| correspond to wider margins, while larger values of ||w|| correspond to narrower margins.

Solving this optimization problem results in finding the optimal weight vector w and bias b that define the hyperplane with the maximum margin while respecting the classification constraints.

Q3. What is the kernel trick in SVM?


Answer(Q3):

The "kernel trick" is a concept in machine learning, specifically in the context of Support Vector Machines (SVMs), that allows SVMs to work efficiently and effectively in higher-dimensional feature spaces without explicitly computing the transformations. It's a mathematical technique that allows you to implicitly compute the dot products between data points in a higher-dimensional space without actually performing the transformation.

In traditional linear SVMs, the algorithm tries to find a hyperplane that best separates the data points of different classes. However, some datasets might not be linearly separable in their original feature space. The kernel trick allows SVMs to effectively find nonlinear decision boundaries by implicitly mapping the original feature space into a higher-dimensional space where the data might be separable. This is accomplished by replacing the standard dot product between feature vectors with a function called a "kernel function."

The general idea is that instead of explicitly transforming the data points into the higher-dimensional space, you define a kernel function \(K(x, x')\) that computes the dot product between two data points \(x\) and \(x'\) in the higher-dimensional space, without actually having to compute the transformation explicitly. This kernel function essentially encapsulates the relationship between the original feature space and the higher-dimensional space.




By using the kernel trick, the SVM can operate in the higher-dimensional space without needing to compute the actual transformation. This leads to more efficient computation and enables the SVM to capture complex decision boundaries that might not be achievable in the original feature space.

To sum up, the kernel trick is a fundamental concept that allows SVMs to work with higher-dimensional feature spaces implicitly, making them capable of capturing nonlinear patterns and enhancing their predictive power on a variety of datasets.

Q4. What is the role of support vectors in SVM Explain with example

Answer(Q4):

Support vectors play a crucial role in Support Vector Machines (SVMs) as they are the data points that lie closest to the decision boundary (hyperplane) and influence its position. They "support" the definition of the hyperplane, and the SVM algorithm focuses on finding the optimal hyperplane that maximizes the margin while maintaining a correct classification of these support vectors.

Let's illustrate this with an example:

Imagine you have a dataset with two classes, "red" and "blue," and the data points are plotted on a 2D plane. Here's a simple example with three data points from each class:


Red:   o     o

           o
            
Blue:     o   o



In this example, let's assume the red class is labeled as "+1" and the blue class as "-1."

The SVM's objective is to find a hyperplane that separates these two classes with the maximum margin. In this case, the hyperplane is a line, and the margin is the distance between the closest data points of the two classes to the hyperplane.

Now, among the data points, the ones that are closest to the hyperplane are the most influential in determining its position. These data points are the support vectors.

In the given example, the three support vectors would be:
1. The red point on the leftmost side of the red class.
2. The blue point on the rightmost side of the blue class.
3. The blue point that is slightly closer to the red class than the other blue point.

The position of these support vectors and their distances from the hyperplane directly influence how the hyperplane is positioned. The SVM's objective is to maximize the margin while ensuring that all support vectors are correctly classified.

In some cases, the other data points, which are not support vectors, might not affect the hyperplane's position if they are far from it. This is an advantage of SVMs—they focus on the most relevant data points, making the algorithm more resistant to outliers.

In summary, support vectors are the critical data points that determine the position of the decision boundary in SVM. They are the ones closest to the hyperplane and are instrumental in maximizing the margin while correctly classifying the data.

Q6. SVM Implementation through Iris dataset.

~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing set

~ Train a linear SVM classifier on the training set and predict the labels for the testing set

~ Compute the accuracy of the model on the testing set

~ Plot the decision boundaries of the trained model using two of the features

~ Try different values of the regularisation parameter and see how it affects the performance of the model.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Selecting only the first two features for visualization
y = iris.target

# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a linear SVM classifier
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X_train, y_train)

# Predict labels for the testing set
y_pred = svm_classifier.predict(X_test)

# Compute the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot decision boundaries
h = .02  # Step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('SVM Decision Boundaries')
plt.show()

# Experiment with different values of the regularization parameter (C)
C_values = [0.01, 0.1, 1, 10, 100]
for C in C_values:
    # Train a linear SVM classifier
    svm_classifier = SVC(kernel='linear', C=C)
    svm_classifier.fit(X_train, y_train)

    # Predict labels for the testing set
    y_pred = svm_classifier.predict(X_test)

    # Compute the accuracy of the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy (C = {C}):", accuracy)
```

Bonus task: Implement a linear SVM classifier from scratch using Python and compare its performance with the scikit-learn implementation.


Answer(Bonus task):



```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np


# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a linear SVM classifier from scikit-learn
scikit_classifier = SVC(kernel='linear')
scikit_classifier.fit(X_train, y_train)

# Predict labels using scikit-learn classifier
scikit_pred = scikit_classifier.predict(X_test)


class SVM:
    def __init__(self, learning_rate=0.01, max_iter=1000, C=1):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.C = C

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.max_iter):
            for idx, x_i in enumerate(X):
                condition = y[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.learning_rate * (2 * self.C * self.w)
                else:
                    self.w -= self.learning_rate * (2 * self.C * self.w - np.dot(x_i, y[idx]))
                    self.b -= self.learning_rate * y[idx]

    def predict(self, X):
        linear_output = np.dot(X, self.w) - self.b
        return np.sign(linear_output)

# Implement and train trail SVM classifier from scratch
trail_svm = SVM(learning_rate=0.01, max_iter=1000, C=1)
trail_svm.fit(X_train, y_train)

# Predict labels using trail own classifier
trail_pred = trail_svm.predict(X_test)


# Compare the performance using accuracy
scikit_accuracy = accuracy_score(y_test, scikit_pred)
trail_accuracy = accuracy_score(y_test, trail_pred)

print("Scikit-learn Accuracy:", scikit_accuracy)
print("trail  Implementation Accuracy:", trail_accuracy)
```
